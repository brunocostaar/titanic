Agora, vamos usar regresso logstica para tentar entender quem sobreviveu no Titanic

```python
formula = 'Survived ~ C(Pclass) + C(Sex) + Age + SibSp + C(Embarked)'
# frmula do modelo
# aqui, o sinal ~  um sinal =, e as caractersticas do nosso conjunto de dados
# so escritas como uma frmula para prever a sobrevivncia. A funo C() permite que nossa
# regresso saiba que essas variveis so categricas.
# Ref: http://patsy.readthedocs.org/en/latest/formulas.html
results = {}
# cria um dicionrio de resultados para armazenar nossos resultados de regresso para facilitar a anlise posterior
```

```python
y,X = dmatrices(formula, data=titanic_data, return_type='dataframe')
#cria um dataframe de regresso utilizando as funes dmatrices
model = sm.Logit(y, X)
res = model.fit()
results['Logit'] = [res, formula]
res.summary()
```

```python
    # Plot Predictions Vs Actual
plt.figure(figsize=(18,4));
# generate predictions from our fitted model
ypred = res.predict(X)
plt.plot(X.index, ypred, 'bo', X.index, y, 'mo', alpha=.25);
plt.grid(color='white', linestyle='dashed')
plt.title('Predies do Logit, Azul: \nValores reais: Rosa');
```

```python
#Resduos
plt.figure(figsize=(18,4));
plt.plot(res.resid_dev, 'r-')
plt.grid(color='white', linestyle='dashed')
plt.title('Resduos do Logit');
```

```python
a = .2
fig = plt.figure(figsize=(18,4));
kde_res = KDEUnivariate(res.predict())
kde_res.fit()
plt.plot(kde_res.support,kde_res.density)
plt.fill_between(kde_res.support,kde_res.density, alpha=a)
plt.title("Distribuio de nossas predies:")
```
